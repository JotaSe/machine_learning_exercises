# Exploring and Transforming Data

```python
import pandas as pd
# Create a dataframe with train data
df = pd.read_csv("data/cs-training.csv",
                     sep=',',
                     header=0)

# Print the first 5 rows of the dataframe.
df.head(5)
```

```python
df.describe()
```

```python
# Fill missing data
df['NumberOfDependents'] = df['NumberOfDependents'].fillna(df['NumberOfDependents'].median())
df['MonthlyIncome'] = df['MonthlyIncome'].fillna(df['MonthlyIncome'].median())
df=df.replace(0,df['age'].median())
df.describe()
```

```python
data = df.drop(
    df.columns[0], 
    axis=1)
###
### Convert Data Into List Of Dict Records
###
data = data.to_dict(orient='records')
```

# Feature extraction

```python
###
### Seperate Target and Outcome Features
###
from sklearn.feature_extraction import DictVectorizer
from pandas import DataFrame
vec = DictVectorizer()
```

```python
df_data = vec.fit_transform(data).toarray()
feature_names = vec.get_feature_names()
df_data = DataFrame(
    df_data,
columns=feature_names)
df_data
```

```python
outcome_feature = df_data['SeriousDlqin2yrs']
target_features = df_data.drop('SeriousDlqin2yrs', axis=1)
```

```python
outcome_feature
```

```python
target_features
```

# Generating Training and Testing Set

```python
###
### Generate Training and Testing Set 
###
from sklearn import cross_validation

"""
    X_1: independent (target) variables for first data set
    Y_1: dependent (outcome) variable for first data set
    X_2: independent (target) variables for the second data set
    Y_2: dependent (outcome) variable for the second data set
"""
X_1, X_2, Y_1, Y_2 = cross_validation.train_test_split(
target_features, outcome_feature, test_size=0.5, random_state=0)
```

# Evaluating Classifier models



```python
from sklearn.metrics import confusion_matrix

def evaluate(clf):
    output = clf.predict(X_2)

    matrix = confusion_matrix(output, Y_2)
    score = clf.score(X_2, Y_2)
    print "accuracy: {0}".format(score.mean())
    print matrix
```

# Naive Bayes

```python
# define classifier                            
from sklearn.naive_bayes import GaussianNB
clf_nb = GaussianNB()
```

```python
# Train Classifier 
clf_nb.fit(X_1,Y_1)
```

```python
# Validate classifier
evaluate(clf_nb)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
clf_rf = RandomForestClassifier(random_state=1, n_estimators=50, min_samples_split=8, min_samples_leaf=4)
clf_rf.fit(X_1,Y_1)
```

```python
evaluate(clf_rf)
```
